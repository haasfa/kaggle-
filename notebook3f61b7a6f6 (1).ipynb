```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

from keras.models import Sequential
from keras.layers import Dense ,BatchNormalization
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from tensorflow.keras.optimizers import Adam, SGD
from matplotlib import pyplot as plt
import numpy

```

```python
column = ['Age', 'Sex', 'CP', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'Slop', 'CA', 'Thal', 'target'] 
# Load the data
df = pd.read_csv('../input/heartdata/heart.csv')
df.head()
```

```python
# Some basic statistical analysis about the data
df.describe()
```

```python
#Define feature set
X = df.drop('target', axis=1)

#Define diagnosis as integer: malignant (1) or benign (0)
y = (df['target'])
```

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# code for RF
from sklearn.ensemble import RandomForestClassifier
RANDOM_STATE = 123

model3 = RandomForestClassifier(n_estimators=100,random_state=RANDOM_STATE)
model3.fit(X_train,y_train)

rounded_predictions3 = model3.predict(X_test)
predictions3 = model3.predict_proba(X_test)
y_pred3 = rounded_predictions3

import sklearn.metrics as metrics
scores3=metrics.accuracy_score(y_test,y_pred3)
print("RF",scores3)

print(metrics.confusion_matrix(y_test,y_pred3))
print(metrics.classification_report(y_test,y_pred3)) 
```
